```python
from attackcti import attack_client
import os
import logging

logging.getLogger('taxii2client').setLevel(logging.CRITICAL)
lift = attack_client()
```

```python
techniques_used_by_groups = lift.get_techniques_used_by_all_groups()
techniques_used_by_groups[0]
```

```python
import pandas as pd
apt_data = []

for tech in techniques_used_by_groups:
  data = {}
  data['tech_name'] = tech['technique']
  data['tech_id'] = tech['technique_id']
  data['description'] = tech['relationship_description']
  data['apt_name'] = tech['name']
  data['apt_aliases'] = tech['aliases']
  data['group_id'] = tech['external_references'][0]['external_id']
  apt_data.append(data)

df = pd.DataFrame(apt_data)

df
```
